# Factory Data Machine Learning

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
from sklearn.impute import SimpleImputer
from sklearn.impute import KNNImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import learning_curve, StratifiedKFold
import matplotlib.pyplot as plt
import pickle
import warnings
warnings.filterwarnings("ignore")
```

### There is a very large amount of data where the machine is of low quality, so using a normal imputer may not be good due to the imbalance. So we will use KNN Imputer instead

## 4. Data Cleaning

```python
process_t = ['Process T (C)'] 
rotation_speed = ['Rotation Speed (rpm)']
quality = ['Quality']
others = ['Torque (Nm)', 'Tool Wear (min)', 'Ambient T (C)']

process_t_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy = "median")),
    ('scaler', StandardScaler())
])

rotation_speed_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy = "mean")),
    ('scaler', StandardScaler())
])

quality_transformer = Pipeline(steps=[
    ('ordinal', OrdinalEncoder()), # The encoder is done first as KNNImputer only works with numeric values
    ('imputer', KNNImputer(n_neighbors=5))
])

others_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

preprocessor = ColumnTransformer(
    transformers=[
        ('process t', process_t_transformer, process_t),
        ('rotation speed', rotation_speed_transformer, rotation_speed),
        ('quality', quality_transformer, quality),
        ('others', others_transformer, others)
    ])
```

### We use Standard Scaler for all the numeric values to standardise them to prevent any single feature from dominating the learning process due to its larger magnitude. Ordinal Encoder is used for "Quality" as its values are ordinal.

### We are putting the imputations into a preprocessor as the data should be split into the training and test data before imputing. This is to reduce data leakage where the test data is being changed, which could lead to the resultss becoming biased


## 5. Data Preprocessing

```python
from imblearn.over_sampling import SMOTE

X = df.drop(columns=['Unique ID', 'Product ID', 'Machine Status'])
y = df['Machine Status']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

X_train_transformed = preprocessor.fit_transform(X_train)
X_test_transformed = preprocessor.transform(X_test)

smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_transformed, y_train)

print(f"Dimensions of original training set: {X_train.shape}")
print(f"Dimensions of resampled training set: {X_train_resampled.shape}")
print(f"Dimensions of test set: {X_test.shape}")

print("\nOriginal training set class distribution:")
print(y_train.value_counts())

print("\nTest set class distribution:")
print(y_test.value_counts())

print("\nResampled training set class distribution:")
print(y_train_resampled.value_counts())
```

### We used SMOTE to generate synthetic samples for the minority class to make the data balanced. This reduces the models' biasness, improving its generalisation and performance on the minority class

## 6. Model Training and Evaluation

### I selected the 4 models that was used during classes for Classification and a Dummy Classifier as a baseline

```python
models = {
    'Dummy Classifier': DummyClassifier(),
    'Logistic Regression': LogisticRegression(max_iter=500),
    'Support Vector Machine': SVC(probability=True),
    'Naive Bayes': GaussianNB(),
    'Random Forest': RandomForestClassifier()
}

for name, model in models.items():
    model.fit(X_train_resampled, y_train_resampled)
    y_pred = model.predict(X_test_transformed)
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='weighted')
    roc_auc = roc_auc_score(y_test, model.predict_proba(X_test_transformed)[:, 1])
    print(f'Model {name}')
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    print("")
    print(f'Model {name} prediction')
    print(f'Accuracy: {accuracy*100:.1f}%')
    print(f'f1-score: {f1*100:.1f}%')
    print(f"ROC AUC Score: {roc_auc*100:.1f}%")
    print("")
```

### From the results, we can see that the Dummy Classifier has a high accuracy and f1 score. However, this is not accuracte as the test data is very imbalanced. As the Dummy Classifier, uses the 'prior' strategy by default, which predicts the majority class for all the samples, as seen in the confusion maxtrix. We can see that the ROC AUC score for the Dummy Classifier is bad, which shows it can't discriminate between positive and negative classes and is randomly guessing. Logistic Regression and Naive Bayes have good scores, while Support Vector Machine and Random Forest has excellent scores. 

## 7. Cross Validation

#### We are doing cross validation to make sure that the model evaluation was accurate or not

```python
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42) # StratifiedKFold over KFold as the test dataset is imbalanced

for name, model in models.items():
    accuracy_scores = cross_val_score(model, X_train_resampled, y_train_resampled, cv=kf)
    f1_scores = cross_val_score(model, X_train_resampled, y_train_resampled, cv=kf, scoring="f1")
    roc_auc_scores = cross_val_score(model, X_train_resampled, y_train_resampled, cv=kf, scoring="roc_auc")
    print(f'Model {name}')
    print(f'Mean Accuracy: {accuracy_scores.mean():.3f}, Std Accuracy: {accuracy_scores.std():.3f}')
    print(f'Mean F1 score: {f1_scores.mean():.3f}, Std F1 score: {f1_scores.std():.3f}')
    print(f'Mean ROC AUC score: {roc_auc_scores.mean():.3f}, Std ROC AUC score: {roc_auc_scores.std():.3f}')
    print("")
```

### Cross Validation again shows that the Dummy Classfier is not good, which is expected. The cross validation scores for the rest of the models are similar, so we will take the two best model, Support Vector Machine and Random Forest, to check for overfitting and underfitting.

### Compared to the Dummy Classifier, the other models performed much better

## 8. Overfitting Check

```python
def plot_learning_curve(model, X, y, title):
    train_sizes, train_scores, test_scores = learning_curve(
        model, X, y, cv=kf, scoring='accuracy', train_sizes=np.linspace(0.05, 1.0, 20), n_jobs=-1
    )

    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)

    plt.figure()
    plt.plot(train_sizes, train_mean, label='Training Score', color='blue')
    plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.1, color='blue')
    plt.plot(train_sizes, test_mean, label='Test Score', color='green')
    plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, alpha=0.1, color='green')
    plt.title(title)
    plt.xlabel('Training Set Size')
    plt.ylabel('Accuracy')
    plt.legend(loc='best')
    plt.grid()
    plt.show()

print("Learning curve for Support Vector Machine")
plot_learning_curve(SVC(), X_train_resampled, y_train_resampled, 'Learning Curve: Support Vector Machine')
print("Learning curve for Random Forest")
plot_learning_curve(RandomForestClassifier(), X_train_resampled, y_train_resampled, 'Learning Curve: Random Forest')
```

### We can see that both the model are not overfitting, so we will take the better model, Random Forest to try and improve it with hyperparameters.

## 9. Hyperparameters Tuning

```python
rf_param_grid = {
    'n_estimators': [100, 200, 300, 500],
    'max_depth': [None, 10, 20, 50],
    'min_samples_split': [2, 5, 10, 15],
    'min_samples_leaf': [1, 2, 4, 10],
    'max_features': ['sqrt', 'log2', None]
}

rf = RandomForestClassifier()

rf_grid = RandomizedSearchCV(
                        estimator=rf,
                        param_distributions=rf_param_grid,
                        cv=kf,
                        n_iter=100,
                        scoring = {'f1_weighted': 'f1_weighted', 'roc_auc': 'roc_auc'},
                        refit='f1_weighted',
                        n_jobs=-1,
                        verbose=1
                        )

rf_grid.fit(X_train_resampled, y_train_resampled)
```

### I used RandomizedSearchCV as I do not need a exhausive search, as after testing both, they gave similar results, with RandomSearchCV being much faster. I chose hyperparameters for Random Forest to try and control model complexity, prevent overfitting, and optimize performance to get a well-balanced model.

```python
best_rf = rf_grid.best_estimator_
best_rf.fit(X_train_resampled, y_train_resampled)

y_pred = best_rf.predict(X_test_transformed)
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
accuracy = 100 * accuracy_score(y_test, y_pred)
f1 = 100 * f1_score(y_test, y_pred, average='weighted')
print('Random Forest after hyperparameters tuning performance')
print(f'Accuracy = {accuracy:.1f}%.')
print(f'f1 score = {f1:.1f}%.')
```

### We can see an improvement in the predictions from the Confusion Matrix, having less false positives and false negatives.

```python
best_rf_features = best_rf.feature_importances_

importance_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': best_rf_features
}).sort_values(by='Importance', ascending=False)

# Plot the feature importances
plt.figure(figsize=(10, 6))
plt.barh(importance_df['Feature'], importance_df['Importance'])
plt.xlabel('Feature Importance')
plt.ylabel('Feature')
plt.title('Random Forest Feature Importance')
plt.gca().invert_yaxis()
plt.show()
```

### From this, we can see that Rotation Speed, Ambient Temperature, and Torque are quite important features, which make sense as these features are related to the power of the machine, which could lead to the machine deteriorating faster, causing it to be abnormal.
